## Задание 2 (20 баллов)

Соберите по каждому студенту среднюю оценку за предмет (агрегация по student_id + class_id) по датасету grades в БД sample_training. Выведите студентов в порядке средней оценки по убыванию, выведите только первые 100 студентов

```python
collection_grades = db_sample_training["grades"]
# for i in collection_grades.find():
#     print(i)
    
grades_agg = collection_grades.aggregate(
    [
        {
            "$unwind": "$scores"
        },
        
        {
            "$group": {
                "_id": ["$student_id", "$class_id"],
                "avg_score": {"$avg": "$scores.score"}
            }
        },
        
        {
            "$sort": {
                "avg_score": pymongo.DESCENDING
            }
        },
        
        {
            "$limit": 100
        }
    ]
)


for i in grades_agg:
    print(i["_id"][0])

# студенты могут повторяться?
```

```python
tags_agg = collection_posts.aggregate(
    [
        {
            "$unwind": "$tags"
        },
        
        {
            "$group": {
                "_id": "$tags",
                "posts_cnt": {"$sum": 1}
            }
        },
        
        {
           "$match": {
              "posts_cnt": {"$eq": 1}
           }
        }
    ]
)


for tag in tags_agg:
    tag_name = tag["_id"]
    print(f"Text with tag {tag_name}:")
    for post in collection_posts.aggregate(
        [
            {
                "$unwind": "$tags"
            },
            
            {"$match": 
                 {
                   "tags": {"$eq": tag_name}
                 }
            }
        ]
    ):
        print(post["body"])
```

## Задание 4 (20 баллов)

Для ресторанов из датасета restaurants БД sample_restaurants выведите самую часто встречающуюся оценку (grade), поставленные в 2014 году. Если у ресторана таких оценок нет, то она не должна быть выведена. Для агрегации используйте name, а не id

Получившийся результат добавьте в дасет test БД test

```python
import datetime

bd_sample_restaurants = cluster.sample_restaurants
collection_restaurants = bd_sample_restaurants.restaurants
# # for i in collection_restaurants.find():
# #     print(i)

restaurants_agg = collection_restaurants.aggregate(
    [
        {
            "$unwind": "$grades"
        },
        {
            "$match": {
                "grades.date": 
                    {"$gte": datetime.datetime(2014, 1, 1, 0, 0),
                    "$lt": datetime.datetime(2015, 1, 1, 0, 0)}
            }
        },
        {
            "$group": {
                "_id": ["$name", "$grades.grade"],
                "grades_cnt": {"$sum": 1}
            }
        },
        {
            "$sort": {
                "_id": 1,
                "grades_cnt": -1
            }
        },
        {
            "$group": {
                "_id": {"$arrayElemAt": ["$_id", 0]},
                "most_frequent_grade": {"$first": {"$arrayElemAt": ["$_id", 1]}},
            }
        },
        {
            "$project": {
                "_id": 0,
                "name": "$_id",
                "most_frequent_grade": 1
            }
        }
    ]
)

for i in restaurants_agg:
    print(i)

db_test = cluster.test
collection_test = db_test.test
# cluster.test.test.delete_many({})

collection_test.insert_many(restaurants_agg)
```

## Задание 5 (30 баллов)

По БД sample_analytics выведите по каждому customer_id (выведите также username аккаунта):

* Сумма лимитов у пользователя (сумма по всем его аккаунтам)

* Общее число транзакций пользователя (сумма по всем аккаунтам)

* (на еще +20 баллов) добавить сумму всех пополнений и трат по транзакциям

```python
cluster.list_database_names()
db_sample_analytics = cluster["sample_analytics"]
db_sample_analytics.list_collection_names()
collection_customers = db_sample_analytics["customers"]
collection_accounts = db_sample_analytics["accounts"]
collection_transactions = db_sample_analytics["transactions"]

# collection_transactions.find_one()
# rows = collection_zips.find()
# for i in rows:
#     print(i)

agg = collection_customers.aggregate(
    [
        { "$unwind": "$accounts" },
        {
            "$lookup": {
              "from": "accounts",
              "localField": "accounts",
              "foreignField": "account_id",
              "as": "account_details"
            }
        },
        
        {
            "$lookup": {
              "from": "transactions",
              "localField": "accounts",
              "foreignField": "account_id",
              "as": "transactions_details"
            }
        },
        
        {
            "$project": {
                "_id": 1,
                "username": 1,
                "account": "$accounts",
                "account_limit": {"$arrayElemAt": ["$account_details.limit", 0]},
                "cnt_transactions": {"$arrayElemAt": ["$transactions_details.transaction_count", 0]}
            }
        },
        
        {
            "$group": {
                "_id": ["$_id", "$username"],
                "sum_limit": {"$sum" : "$account_limit"},
                "sum_transactions_cnt": {"$sum" : "$cnt_transactions"}
                
            }
        },
    ]
)

for k in agg:
    print(k)
```
